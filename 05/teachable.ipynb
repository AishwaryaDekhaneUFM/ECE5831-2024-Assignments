Explanation of the Image Classification Script

This script demonstrates how to perform image classification using a pre-trained Keras model. The steps involved include loading the model and labels, preprocessing the input image, making a prediction, and displaying the predicted class and confidence score. Here’s a step-by-step breakdown of what the code is doing:

1. **Import Necessary Libraries:**
   - The script begins by importing essential libraries. These include TensorFlow (required for Keras to work), Pillow (for image processing), and NumPy (for handling arrays and numerical operations).

2. **Disable Scientific Notation:**
   - To ensure that the output numbers are easy to read, scientific notation is disabled in NumPy’s settings.

3. **Load the Pre-Trained Model:**
   - The pre-trained Keras model is loaded from a specified file. This model has been trained on a particular dataset and can make predictions based on the input images.

4. **Load Class Labels:**
   - The script reads class labels from a text file. These labels correspond to the classes that the model can predict. Each line in the file represents a different class.

5. **Prepare Input Data:**
   - An array of the correct shape is created to feed into the model. This array will hold the preprocessed image data.
   - The input image is loaded and converted to RGB format. This ensures that the image has the correct color channels for processing.
   - The image is resized to 224x224 pixels. This resizing ensures that the image dimensions match the input dimensions expected by the model.
   - The resized image is converted into a NumPy array and normalized. Normalization scales the pixel values to a range suitable for the model, typically from -1 to 1.

6. **Load Image Data into Array:**
   - The normalized image data is placed into the previously created array, making it ready to be fed into the Keras model for prediction.

7. **Make Predictions:**
   - The model makes a prediction on the input image. The output is an array of probabilities corresponding to each class.
   - The index of the highest probability is found, which indicates the predicted class.
   - The corresponding class label and confidence score (probability of the predicted class) are extracted from the predictions.

8. **Display Predictions:**
   - The predicted class label and confidence score are printed out. The class label indicates what the model predicts the image to be, and the confidence score indicates how sure the model is about its prediction.

This script allows you to take any input image, preprocess it appropriately, and use a pre-trained Keras model to classify the image into one of the recognized classes. The predicted class and the confidence score are displayed as output, providing insights into the model's prediction.





```python
from keras.models import load_model  # TensorFlow is required for Keras to work
from PIL import Image, ImageOps  # Install pillow instead of PIL
import numpy as np

```

```python
# Disable scientific notation for clarity
np.set_printoptions(suppress=True)

# Load the model
model = load_model("/content/keras_model.h5", compile=False)

# Load the labels
class_names = open("/content/labels.txt", "r").readlines()

# Create the array of the right shape to feed into the keras model
# The 'length' or number of images you can put into the array is
# determined by the first position in the shape tuple, in this case 1
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

# Replace this with the path to your image
image = Image.open("/content/103.jpg").convert("RGB")

import matplotlib.pyplot as plt

plt.imshow(image)
```

Sure! Here's an explanation of what each part of this segment of the script is doing:

1. **Resizing and Cropping the Image:**
   - The script resizes the image to ensure it is at least 224x224 pixels in dimensions. This resizing is done using a high-quality resizing algorithm (Lanczos resampling), which helps in maintaining image quality during the resizing process. After resizing, the image is cropped from the center to ensure it fits the exact dimensions required by the model (224x224 pixels).

2. **Converting the Image to a NumPy Array:**
   - Once the image is resized and cropped, it is converted into a NumPy array. This conversion is necessary because the Keras model expects the input data to be in array format.

3. **Normalizing the Image:**
   - The image array is then normalized. Normalization in this context involves scaling the pixel values to a range that the model can efficiently process. Specifically, pixel values are scaled from their original range (0-255) to a new range of -1 to 1. This is achieved by converting the pixel values to floating-point numbers and adjusting the range accordingly.

4. **Preparing the Image for the Model:**
   - The normalized image array is then loaded into the previously created array, setting it as the first (and only) item in that array. This step ensures that the image data is in the correct format and ready to be fed into the Keras model for prediction.

5. **Making Predictions:**
   - The model is then used to make a prediction on the input image. It outputs probabilities for each possible class, indicating how likely it thinks the image belongs to each class.

6. **Interpreting the Predictions:**
   - The script identifies the class with the highest probability (maximum probability) as the predicted class. It does this by finding the index of the highest probability value in the model's predictions. This index corresponds to the predicted class label.

7. **Fetching the Class Name and Confidence Score:**
   - The class label is retrieved from the list of class names using the identified index. Additionally, the confidence score, which is the probability of the predicted class, is extracted from the model’s output.

8. **Displaying the Results:**
   - Finally, the script prints the predicted class name and the confidence score. The predicted class name is stripped of any leading characters (typically spaces or labeling characters) and displayed alongside the confidence score, indicating how confident the model is in its prediction.

This section of the script handles the critical steps of image preprocessing, prediction, and result interpretation, allowing the user to see which class the model predicts for the input image and how confident it is in that prediction.

```python
# resizing the image to be at least 224x224 and then cropping from the center
size = (224, 224)
image = ImageOps.fit(image, size, Image.Resampling.LANCZOS)

# turn the image into a numpy array
image_array = np.asarray(image)

# Normalize the image
normalized_image_array = (image_array.astype(np.float32) / 127.5) - 1

# Load the image into the array
data[0] = normalized_image_array

# Predicts the model
prediction = model.predict(data)
index = np.argmax(prediction)
class_name = class_names[index]
confidence_score = prediction[0][index]

# Print prediction and confidence score
print("Class:", class_name[2:], end="")
print("Confidence Score:", confidence_score)
```
