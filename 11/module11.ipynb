## IMDB Class

```python
from imdb import IMDB

# Initialize and prepare data
imdb_instance = IMDB(num_words=10000)
imdb_instance.prepare_data()
```

```python
# Build and train the model
imdb_instance.build_model()
imdb_instance.train(epochs=20, batch_size=512)
```

```python
# Plot loss results
imdb_instance.plot_loss()
```

```python
# Plot training results
imdb_instance.plot_accuracy()
```

```python
# Evaluate the model
imdb_instance.evaluate()
```

## Reuters Class

```python
from reuters import Reuters
```

```python
# Initialize and prepare data
reuters_instance = Reuters(num_words=10000, num_classes=46)
reuters_instance.prepare_data()
```

```python
# Build and train the model
reuters_instance.build_model()
reuters_instance.train(epochs=20, batch_size=512)
```

```python
# Plot loss results
reuters_instance.plot_loss()
```

```python
# Plot training results
reuters_instance.plot_accuracy()
```

```python
# Evaluate the model
reuters_instance.evaluate()
```
